# scaling numeric features 
- Absolute Maximum Scaling Find the absolute maximum value of the feature in the dataset
Divide all the values in the column by that maximum value sensitive to outliers



- Min-Max Scaling:  min-max you will subtract the minimum value in the dataset with all the values and then divide this by the range of the dataset(maximum-minimum)  X_new = X - X_min / X_max - X_min



- Normalization: In scaling, you are changing the range of your data while in normalization you are changing the shape of the distribution of your data.  X_new = X - X_mean / X_max - X_min



- Standardization: In standardization, we calculate the z-value for each of the data points and replaces those with these values - This will make sure that all the features are centred around the mean value with a standard deviation value of 1. This is the best to use if your feature is normally distributed like salary or age  
X_new =  X - X_mean/ sd 






- Robust Scaling: In this method, you need to subtract all the data points with the median value and then divide it by the Inter Quartile Range(IQR) value.



```python
from sklearn.preprocessing import MinMaxScaler
```

```python
# fit all the colums here 
scaler = MinMaxScaler().fit(df[numeric_cols])
```

```python
train_inputs[numeric_cols] = scaler.transform(train_inputs[numeric_cols])
val_inputs[numeric_cols] = scaler.transform(val_inputs[numeric_cols])
test_inputs[numeric_cols] = scaler.transform(test_inputs[numeric_cols])
```

# Encoding Categorical Variables

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
# fit all the categories here 
encoder = OneHotEncoder(sparse=False, handle_unknown='ignore').fit(df[categorical_cols])
```

```python
encoded_cols = list(encoder.get_feature_names_out(categorical_cols))
encoded_cols
```

```python
train_inputs[encoded_cols] = encoder.transform(train_inputs[categorical_cols])
val_inputs[encoded_cols] = encoder.transform(val_inputs[categorical_cols])
test_inputs[encoded_cols] = encoder.transform(test_inputs[categorical_cols])
```

```python
train_inputs
```

```python
# grab the numeric and envcoded columns
X_train = train_inputs[numeric_cols + encoded_cols]
X_val = val_inputs[numeric_cols + encoded_cols]
X_test = test_inputs[numeric_cols + encoded_cols]
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier(random_state=42)
```

```python
%%time
model.fit(X_train, train_targets)
```

```python
 y_pred = model.predict(X_train)
```
